```python
#librerias
from tensorflow.keras.models import Sequential #red secuencial
from tensorflow.keras.layers import Dense # manejo de capas
import numpy #matematica
#aleatorio poner siempre la misma semilla
numpy.random.seed(9)
```

```python
#dataset datos.csv = es un estudio real de gente que se le desarrollo la diabetes en 5 anios
#columna 1=embarazos, 2= concentracion de glucosa. 3= presion arterial, 4= graso del pliegue, 5= insluina
#6= masa corporal, 7= antecedentes diabeticos, 8= edad, 9= es el resultado de 5 anios 1 se desarrolla 0 no
dataset=numpy.loadtxt("C:/Users/shere/Downloads/datos.csv",delimiter=",")
#las x(entradas) y(salidas)
x=dataset[:,0:8]
y=dataset[:,8]
```

```python
#crear el modelo
model=Sequential() #una red secuencial capa de entrada pasa capa oculta hasta llegar a la capa de salida
#voy a definir las neuronas de entrada, funcion de activacion
model.add(Dense(12,input_dim=8,activation='relu'))
#capa oculta
model.add(Dense(8,activation='relu'))
#capa de salida la sigmoidal va de 0 a 1 y perfectamente encaja de abajo de .5 es 0 y arriba de .5 es 1
model.add(Dense(1,activation='sigmoid'))
```

```python
#compilar la red, identificar aspectos como optimizador = gradiente descendente que ajusta los pesos Wi, loss= para ir mejorando
# minimizar errores cuadrados
model.compile(loss="binary_crossentropy", optimizer="adam",metrics=['accuracy'])
```

```python
#modelar la red, epochs son iteraciones y batch_size son pasadas por iteracion
model.fit(x,y,epochs=150,batch_size=10)
```

```python
#evalua el modelo
scores = model.evaluate(x,y)
print("\n %s: %.2f%%" %(model.metrics_names[1],scores[1]*100))
```

```python
#la prediccion que tienes que hacer siempre es con el modelo
predictions = model.predict(x)
#salida de mis predicciones, redondeo las predicciones .5 a 0 y otro caso a 1
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
